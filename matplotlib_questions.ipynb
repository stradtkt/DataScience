```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Create a figure object called fig using plt.figure

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
```

## Create a figure object and put two axes on it ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,0.2,0.2]

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,.5,.2,.2])
ax1.plot(x, y)
ax2.plot(x, y)
```

## Create the plot below by adding two axes to a figure object at [0,0,1,1] and [.2,.5,.4,.4]

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

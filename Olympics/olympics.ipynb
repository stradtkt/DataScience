```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
o = pd.read_csv('athlete_events.csv')
o.head()
```

```python
usa = o[(o['NOC'] == "USA") & (o['Medal'] == "Gold") & (o["Year"] == 2016)]
usa.head(15)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,5))
sns.countplot(x='Sex', data=usa)
sns.despine()
female = usa[usa['Sex'] == 'F'].count()['ID']
male = usa[usa['Sex'] == 'M'].count()['ID']
print("Male Gold Medal Winners:", male)
print("Female Gold Medal Winners:", female)
```
